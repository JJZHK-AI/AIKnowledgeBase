```python
import itertools
from mxnet import contrib, gluon, image, nd
import math
import numpy as np
import d2lzh as d2l

img = image.imread('data/catdog.jpg')
h, w = img.shape[0:2]
print(h, w)
```

```python
def display_anchors(fmap_w, fmap_h, s):
    fmap = nd.zeros((1, 3, fmap_w, fmap_h))  # 前两维的取值不影响输出结果
    anchors = contrib.nd.MultiBoxPrior(fmap, sizes=s, ratios=[1, 2])
    print(anchors[0])
    bbox_scale = nd.array((w, h, w, h))
    d2l.show_bboxes(d2l.plt.imshow(img.asnumpy()).axes,
                    anchors[0] * bbox_scale)
```

```python
display_anchors(2, 2, [0.3])
```
