奇异值分解(SVD)
====
图片压缩

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# 灰度化读取图片
img = cv2.imread('data/beach.jpg', 0)

# 得到svd
U, S, V = np.linalg.svd(img)

# 得到矩阵的形状
print(U.shape, S.shape, V.shape)

# 以不同component数绘制图像
comps = [638, 600, 500, 400, 300, 200, 100, 50, 10]

plt.figure(figsize = (16, 8))
for i in range(9):
    low_rank = U[:, :comps[i]] @ np.diag(S[:comps[i]]) @ V[:comps[i], :]
    if(i == 0):
        plt.subplot(3, 3, i+1)
        plt.imshow(low_rank, cmap = 'gray')
        plt.axis('off')
        plt.title("Original Image with n_components =" + str(comps[i]))
    else:
        plt.subplot(3, 3, i+1)
        plt.imshow(low_rank, cmap = 'gray')
        plt.axis('off')
        plt.title("n_components =" + str(comps[i]))
```
