导数与变化率
===

# 1.导数的几何解释

## 1.1.导数的引入
问题：如果求解一条曲线y在$(x_0,y_0)$出的切线？

我们知道在高中的知识中，函数在某一点处的切线可以用方程$y-y_0=m(x-x_0)$来表示，对此我们还需要清楚一下两点
$$\begin{cases}
y_0=f(x_0) \\
m=f'(x_0)
\end{cases}$$
其中m叫做斜率(Slop)，它是函数在$x_0$这一点上的导数

## 1.2.导数的定义
$f'(x_0)$它是函数$f$在$x_0$处的导数，也是$y=f(x)$在此点的斜率。

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(-2, 10, 100)
y = np.power(2, x)
plt.figure(figsize=(8,8),facecolor='w')
plt.plot(x,y,color="red",linewidth=2)
slot = np.power(2,4) * np.log(2)
b = -28.6
plt.plot(x, slot * x + b, color="g")
plt.plot(x, 4 * x, color="b")
plt.ylim(-30,30)
plt.text(3.5, 16, 'P', fontsize=20)
plt.text(0, 2, 'D', fontsize=20)
ax = plt.gca()

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.show()
```

我们可以看到在P点处的切线是绿线，蓝线是经过P点的一条割线。那么我们可以从极限的角度来解释导数和切线。如果我们叫D点沿着曲线不停的去靠近P点，那么蓝色的割线也会不断的去拟合绿色的切线，当D和P无限接近时，割线和切线就融合成了一条直线。所以从这个角度看，导数就是极限

## 1.3.符号引入
P点的坐标我们记作$(x_0,y_0)$，D点和P点的水平坐标差我们记作$\Delta{x}$，也叫做$x$的变化量，垂直坐标差我们记作$\Delta{f}$，也叫做$f$的变化量，那么割线PQ的斜率就是$\frac{\Delta{f}}{\Delta{x}}$，那么根据极限的分析，P点切线的斜率就是$m=\lim\limits_{\Delta{x} \to 0 }{\frac{\Delta{f}}{\Delta{x}}}$。

## 1.4.推导简化
P点的坐标我们可以用$x_0,f(x_0)$来表示，那么Q点的坐标我们就可以使用$x_0+\Delta{x},f(x_0+\Delta{x})$来表示了，因此就有:
$$f'(x_0)=m=\lim\limits_{\Delta{x} \to 0}{\frac{f(x_0+\Delta{x}) - f(x_0)}{\Delta{x}}}$$
这是最重要的公式，可以推导出所有其它的公式

### 1.4.1.实例I
有函数$f(x)=\frac{1}{x}$，求在$x_0$处的导数。<br/>
首先带入导数公式有
$$\begin{split}
f'(x_0) &=\lim\limits_{\Delta{x} \to 0}{\frac{f(x_0+\Delta{x}) - f(x_0)}{\Delta{x}}} \Rightarrow \lim\limits_{\Delta{x} \to 0}{\frac{\frac{1}{x_0+\Delta{x}}-\frac{1}{x_0}}{\Delta{x}}} \Rightarrow \lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}}(\frac{1}{x_0+\Delta{x}}-\frac{1}{\Delta{x}}) \\\\
&\Rightarrow \lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}}[\frac{x_0-(x_0+\Delta{x})}{(x_0+\Delta{x})x_0}] \Rightarrow \lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}} \times \frac{-\Delta{x}}{(x_0+\Delta{x})x_0} \Rightarrow \lim\limits_{\Delta{x} \to 0}\frac{-1}{(x_0+\Delta{x})x_0}\\\\
&=\frac{-1}{x_0^2}
\end{split}$$

### 1.4.2.实例II
求由$y=\frac{1}{x}$切线和坐标轴合围而成的三角形的区域面积

```python
from matplotlib import pyplot as plt
import numpy as np

x = np.linspace(0.1, 5, 1000)
y = 1/x
plt.figure(figsize=(8,8), facecolor='w')
plt.plot(x,y,color="red",linewidth=2)
plt.scatter(1,1, color='b')
slot = -1 / 1
b = 2
plt.plot(x, slot * x + b, color="g")
plt.ylim(0,4)
ax = plt.gca()

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.show()
```

切线方程如下:
$$y-y_0=\frac{-1}{x_0^2}(x-x_0)$$
令$y=0$，可以找到切线跟x轴相交的坐标，有
$$\begin{split}
y-y_0&=\frac{-1}{x_0^2}(x-x_0) \\\\
0-\frac{1}{x_0}&=\frac{-1}{x_0^2}(x-x_0) \\\\
&\Rightarrow x=2x_0
\end{split}$$
同样在y轴上的坐标为$y=2y_0$，因此面积为$\frac{1}{2} \times 2x_0 \times 2y_0=2x_0y_0=2$，是个常量

## 1.5.符号总结
$$\begin{split}
y &=f(x) \\\\
\Delta{y} &=\Delta{f} \\\\
f'&=\frac{dy}{dx}=\frac{df}{dx}=\frac{d}{dx}y=\frac{d}{dx}f
\end{split}$$
$f'$是牛顿使用的，后面的是莱布尼茨发明的

## 1.6.实例
有函数$f(x)=x^n, n=1,2,3,...$，求$\frac{d}{dx}x^n$。<br/>
根据定义有
$$\frac{d}{dx}x^n=\lim\limits_{\Delta{x} \to 0}\frac{(x+\Delta{x})^n-x^n}{\Delta{x}}$$
根据二项式定理，我们可以有如下展开
$$\begin{split}
(x+\Delta{x})^n &=(x+\Delta{x}) \times \cdots \times (x+\Delta{x}) \\\\
&\Rightarrow x^n+nx^{n-1}\Delta{x}+Junk
\end{split}$$
$Junk=\mathcal{O}(\Delta{x})^2$含义就是$\Delta{x}$具有更高阶.继续推导
$$\begin{split}
\frac{d}{dx}x^n &=\lim\limits_{\Delta{x} \to 0}\frac{(x+\Delta{x})^n-x^n}{\Delta{x}} \\\\
&=\lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}}[(x+\Delta{x})^n-x^n] \\\\
&=\lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}}[x^n+nx^{n-1}\Delta{x}+\mathcal{O}(\Delta{x})^2-x^n] \\\\
&=\lim\limits_{\Delta{x} \to 0}\frac{1}{\Delta{x}}[nx^{n-1}\Delta{x}+\mathcal{O}(\Delta{x})^2] \\\\
&= \lim\limits_{\Delta{x} \to 0}[nx^{n-1}+\mathcal{O}(\Delta{x})] \\\\
&= nx^{n-1}
\end{split}$$

# 2.导数的物理学解释
对于第二节中的例子来说，如果我们把x轴当做时间，那么我们可以叫$\frac{\Delta{y}}{\Delta{x}}$为变化率，也就是在$\Delta{x}$时间区间上的平均变化率，那么$\frac{dy}{dx}$就是瞬时变化率。在物理学中，如果q表示电荷(charge)，那么$\frac{dq}{dt}$就便是电流(current)；如果s表示距离，那么$\frac{ds}{dt}$就表示速度；假设T表示温度，那么$\frac{dT}{dx}$就叫做温度梯度

## 2.1.重要实例I
假设有一栋80m高的大楼，然后有人从楼顶往下扔南瓜。由于重力加速度，我们可以得到南瓜离地面的距离与时间有如下关系:
$$h=80-5t^2$$
也就是当$t=0$时，$h=80-5 \times o^2=80$，当$t=4$时，$h=80-5 \times 4^2=0$，南瓜正好落地。那么我们可以得到南瓜运行的平均速度，也就是平均变化率为$\frac{\Delta{h}}{\Delta{t}}=\frac{0-80}{4-0}=-20m/sec$,那么当南瓜落地时的瞬间速度，也就是瞬间变化率就是$h'(t)=\frac{d}{dt}h=\frac{d}{dt}(80-5t^2)=-10t$，那么当$t=4$的时候，瞬间速度就是$-40m/sec=-144km/hour$，这是一个相当大的速度$144$迈
