导数的应用-线性近似与二阶近似
===

# 1.公式引入
$$f(x) \approx f(x_0)+f'(x_0)(x-x_0)$$
假设有一条曲线$y=f(x)$,那么它在切点处近似于其切线，也就是说公式的右边是曲线的切线

 ## 1.1.实例
 假设有函数$y=\ln{x}$

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')
x = np.linspace(0, 10, 1000)
y = np.log(x)

plt.plot(x, y, color="r", linewidth=2)
plt.scatter(1, np.log(1), color="b")
plt.text(1, 0, 'D', fontsize=20)

x = np.linspace(-10, 10, 1000)
y = x-1
plt.plot(x, y, color="g", linewidth=2)
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid(True)
```

> $$\begin{split}
&\because f(x)=\ln{x} \therefore f'(x)=\frac{1}{x} \\\\
&set x_0=1, \Rightarrow f(1)=\ln{1}=0, f'(1)=1 \\\\
&\Rightarrow ln(x) \approx 0 + 1 \times (x-1) \\\\
&\Rightarrow ln(x) \approx x-1
\end{split}$$

我们可以看到图上的绿线就是$y=x-1$的图像，在$x$近似于1的时候，$\ln{x}$近似于$x-1$

# 2.公式II
还有一个近似$$\frac{\Delta{f}}{\Delta{x}} \approx f'(x_0)$$

> 证明公式I与公式II是等价的
$$\begin{split}
\Delta{f}&\approx f'(x_0) \times \Delta{x} \\\\
f(x)-f(x_0)&\approx f'(x_0)(x-x_0) \\\\
f(x) &\approx f(x_0)+f'(x_0)(x-x_0)
\end{split}$$

千万记住这两个公式都是在条件$x \approx x_0$的时候才能成立

# 3.常用的线性近似公式
- $x \approx 0$, $\ln{1+x} \approx x$
- $x \approx 0$, $(1+x)^r \approx 1+rx$
- $x \approx 0$, $\sin{x} \approx x$
- $x \approx 0$, $\cos{x} \approx 1$
- $x \approx 0$, $e^x \approx 1+x$

# 4.线性近似的应用
> 求$\ln{1.1}$的值。我们根据线性近似的公式可以推导出当$x \approx 0$时，$\ln{1+x} \approx x$，所以$\ln{1.1} \approx \frac{1}{10}$

> 求函数$\frac{e^{-3x}}{\sqrt{1+x}}$在$x \approx 0$处的线性近似。
$$\begin{split}
\frac{e^{-3x}}{\sqrt{1+x}}&=e^{-3x} \times (1+x)^\frac{-1}{2} \\\\
\frac{e^{-3x}}{\sqrt{1+x}} &\approx (1-3x) \times (1-\frac{1}{2}x) \\\\
&\Rightarrow \frac{3}{2}x^2-\frac{7}{2}x+1
\end{split}$$

我们需要扔掉高阶项$\frac{3}{2}x^2$，最终结果就是$1-\frac{7}{2}x$

# 5.二阶近似
二阶近似是线性近似的扩展，比线性近似更精确
$$
f(x) \approx f(x_0)+f'(x_0)(x-x_0)+\frac{f''(x_0)}{2}(x-x_0)^2
$$

# 6.常用的二阶近似公式
- $x \approx 0$, $\ln{1+x} \approx x-\frac{x^2}{2}$
- $x \approx 0$, $(1+x)^r \approx 1+rx+\frac{r(r-1)}{2}x^2$
- $x \approx 0$, $\sin{x} \approx x$
- $x \approx 0$, $\cos{x} \approx 1-\frac{1}{2}x^2$
- $x \approx 0$, $e^x \approx 1+x+\frac{1}{2}x^2$

# 7.二阶近似的几何意义

```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')
x = np.linspace(-10, 10, 1000)
y = np.cos(x)

plt.plot(x, y, color="r", linewidth=2)
plt.scatter(0, 1, color="b")
plt.text(0, 1, 'D', fontsize=20)

plt.plot(x, np.ones(1000), color="b", linewidth=2)

plt.plot(x, 1-x**2/2, color="g", linewidth=2)

ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.grid(True)
```

图中红线表示$y=\cos{x}$，蓝线是在$x \approx 0$的时候的线性近似，绿线是在$x \approx 0$的时候的二阶近似。可以看出二阶近似更接近原始图像的曲线。

# 8.线性近似和二阶近似的应用

> 通过线性近似来证明$a_k=(1+\frac{1}{k})^k$,当$k \to \infty$时，$a_k \to e$。<br/>
$$\begin{split}
&\forall x \approx 0, \ln{(1+x)} \approx x \\\\
&\Rightarrow \ln{a_k} = k \times \ln{(1+\frac{1}{k})} \approx k \times \frac{1}{k} \approx 1
\end{split}$$
上式成立的条件是$\frac{1}{k} \approx 0$，那么也就是相当于当$k \to \infty$的时候$k \times \ln{(1+\frac{1}{k})} \to 1$，同时也就证明了$\lim\limits_{k \to \infty}(1+\frac{1}{k})^k=e$

上面证明了当$k \to \infty$的时候，$\ln{a_k} \to 1$，也就是说$\ln{a_k}-1 \to 0$，但是我们现在想要知道它趋近于0的速度，也就是变化率，这个时候就需要用到二阶近似
