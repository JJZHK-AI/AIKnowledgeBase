曲线构图与最值问题
===
曲线构图就是利用$f'$以及$f''$来画出f的图像

# 1.一些概念
> 如果$f' > 0$，那么函数递增；如果$f'<0$，那么函数递减；<br/>
> 如果$f''>0$，那么$f'$递增，$f$是凸函数；如果$f''<0$，那么$f'$递减，$f$是凹函数

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')
x = np.linspace(-10, 10, 1000)
y = 3*x- x**3
z = 3-3 ** x**2
plt.plot(x, y, color="r", linewidth=2)
plt.plot(x, z, color="g", linewidth=2)

plt.scatter(1, 2, color='b')
plt.scatter(-1, -2, color='b')
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```

上图我们画出了$f(x)=3x-x^3$以及它的导数$f(x)=3-3x^2$的图像，通过一阶导数和二阶导数来还原原图像的过程中，首先需要找到驻点。
> 1. $f'(x)=3-3x^2=3 \times (1-x) \times (1+x)$
> 2. 分析可以得到如下结果
$$\begin{cases}
x < -1, f'(x)<0\\\
-1 < x < 1, f'(x) > 0\\\
x > 1, f'(x) < 0
\end{cases}$$
> 3. 我们可以知道在$x<-1，x>1$的时候，函数是递减的，$-1 < x < 1$的时候，函数是递增的。这样就可以大概画出原图像了。同时(-1,-2),(1,2)这两个点叫做驻点。驻点不一定是极值点，但极值点一定是驻点
> 4. 二阶导数为$f''(x)=-6x$，可以知道$(0,0)$点左边，图像是凹函数，右边是凸函数，这个点叫做拐点

# 2.最大值和最小值

## 2.1.实例
> 有一条长度为1的绳子，剪成两段，每一段圈成一个正方形，求所能得到的最大面积

设其中一条线段的长度为$x$，那么另外一条线段的长度就是$1-x$，那么两个正方形的边长就分别是$\frac{x}{4},\frac{1-x}{4}$，那么就有面积A为$A=(\frac{x}{4})^2+(\frac{1-x}{4})^2$。求A的最大值，我们就要找到这个图像的驻点.<br/>
> $f'(A) = \frac{d}{dx}A = \frac{x}{8} - \frac{1-x}{8} =0$，$x=\frac{1}{2}$,面积就是$\frac{1}{32}$

我们需要检查边界点，因为这道题目有一个潜在的限制就是$0<x<1$
$$\begin{split}
A(0^+)&=0+(\frac{1}{4})^2=\frac{1}{16}\\\\
A(1^-)&=(\frac{1}{4})^2+0=\frac{1}{16}
\end{split}$$

```python
from matplotlib import pyplot as plt
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')
x = np.linspace(0, 1, 1000)
y = (x/4)**2 + ((1-x)/4)**2

plt.plot(x, y, color="r", linewidth=2)
# plt.plot(x, -6 * x, color='b',linewidth=2)

plt.scatter(0, 1/16, color='b')
plt.scatter(1, 1/16, color='b')
plt.scatter(1/2, 1/32, color='b')
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.xlim(-1,1)
plt.ylim(0,0.1)
plt.show()
```

我们可以看到我们得到的是最小面积，而不是最大面积，因为这个图像是一个凹函数。当只有一个正方形的时候，面积最大，是$\frac{1}{16}$
