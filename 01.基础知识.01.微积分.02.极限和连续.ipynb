极限与连续
===

# 1.极限的一些概念
- 简单极限: $\lim\limits_{x \to 4}\frac{x+3}{x^2+1}=\frac{4+3}{4^2+1}=\frac{7}{17}$
- 复杂极限：导数一般都是比较复杂的极限。$\lim\limits_{x \to x_0}\frac{f(x_0+\Delta{x})-f{x_0}}{x-x_0}$。这种极限不能直接带入，否则就是$\frac{0}{0}$。这种极限需要先做对消操作
- 右极限：$\lim\limits_{x \to x^+}f(x)$。此时在坐标轴上，x位于$x_0$的右侧
- 左极限：$\lim\limits_{x \to x^-}f(x)$。此时在坐标轴上，x位于$x_0$的左侧

## 1.1.重要实例I
$$
f(x)=\begin{cases}
x+1, & x>0 \\
-x+2, & x<0
\end{cases}
$$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')

x = np.linspace(0.01, 10, 1000)
y = x+1
plt.plot(x,y,color="red",linewidth=2)

x = np.linspace(-10, -0.01, 1000)
y = -x+2
plt.plot(x,y,color="b",linewidth=2)
plt.ylim(0,15)
ax = plt.gca()

ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.grid(True)
```

- $\lim\limits_{x \to 0^+}f(x)=\lim\limits_{x \to 0}(x+1)=1$
- $\lim\limits_{x \to 0^-}f(x)=\lim\limits_{x \to 0}(-x+2)=2$

# 2.连续
如果我们说$f$在$x_0$点连续，那么就是说当$x \to x_0$时$f(x)$的极限等于$f(x_0)$。$x_0$点的连续包含了一下内容:
- $\lim\limits_{x \to x_0}f(x)$必须存在，不管是左极限还是有极限，并且左右极限必须相等
- $f(x_0)$必须被定义
- 商量着必须相等，也就是说$\lim\limits_{x \to x_0}f(x)=f(x_0)$。这不就是定义吗？

# 3.间断

## 3.1.跳跃间断
左右极限均存在，但不相等，比如上例

## 3.2.可去间断
左右极限存在且相等，但中间有一个点是没有定义的。以下两个重要的极限都是可去间断
- $\lim\limits_{x \to 0}\frac{sinx}{x}=1$
- $\lim\limits_{x \to 0}\frac{1-cosx}{x}=0$

## 3.3.无穷间断
$f(x)=\frac{1}{x}$就是无穷间断
- $\lim\limits_{x \to 0^+}\frac{1}{x}=\infty$
- $\lim\limits_{x \to 0^-}\frac{1}{x}=-\infty$

它的导数是$\frac{1}{x^2}$，那么它的极限就是$\lim\limits_{x \to 0}\frac{-1}{x^2}=-\infty$

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(16,8), facecolor='w')
plt.subplot(1, 2, 1)
x = np.linspace(-1, 1, 100)
y = 1/x
plt.plot(x,y,color="red",linewidth=2)
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.grid(True)
plt.subplot(1, 2, 2)
x = np.linspace(-1, 1, 100)
y = -1/np.power(x, 2)
plt.plot(x,y,color="red",linewidth=2)

plt.ylim(-100,100)
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))
plt.grid(True)
```

原函数$\frac{1}{x}$是奇函数，求导后的函数变成了偶函数。这是一个定理。奇函数求导后会得到偶函数。
- 奇函数：定义域内关于原点对称的函数。即$f(x)$的定义域内任意一个x，都有$f(-x)=-f(x)$
- 偶函数：定义域内关于y轴对称的函数。即$f(x)$的定义域内任意一个x，都有$f(-x)=f(x)$

## 3.4.另类(丑陋)间断
没有左右极限。比如$y=sin\frac{1}{x}$，趋近于0时，它会不断震荡

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

plt.figure(figsize=(8,8), facecolor='w')

x = np.linspace(-20, 20, 100)
y = np.sin(1/x)
plt.plot(x,y,color="red",linewidth=2)
ax = plt.gca()
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data',0))

plt.ylim(-1,1)
plt.grid(True)
```
